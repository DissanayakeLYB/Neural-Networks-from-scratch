## Linear Regression

- Used to find the linear correspondence of a datapoints.
- When given some x values, we need th model to find the most suitable y for that x value/s.

**Line of best fit** refers to a line through a scatter plot of data points that best expresses tge relationship between those points.

- This can be done on higher dimensions as well.

Linear Regression is mostly used when the datapoints are expected to have a linear relationship. (corelation)
